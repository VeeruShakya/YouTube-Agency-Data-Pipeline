```python
import pandas as pd
from sqlalchemy import create_engine

print("Starting Agency ETL Pipeline...")

# --- PHASE 1: INGESTION & CLEANING ---
df_raw = pd.read_csv('creator_raw_data.csv')
df_clean = df_raw.dropna(subset=['Views'])
df_clean['Sponsor_Payout_USD'] = df_clean['Sponsor_Payout_USD'].fillna(0)
df_clean['Agency_Fee_Percent'] = df_clean['Agency_Fee_Percent'].fillna(0)
df_clean['Sponsor_Brand'] = df_clean['Sponsor_Brand'].fillna('No Sponsor')

# --- PHASE 2: FEATURE ENGINEERING ---
df_clean['Total_Video_Revenue_USD'] = df_clean['Adsense_Revenue_USD'] + df_clean['Sponsor_Payout_USD']
df_clean['Agency_Earnings_USD'] = (df_clean['Sponsor_Payout_USD'] * df_clean['Agency_Fee_Percent']) / 100


df_clean.to_csv('creator_clean_data.csv', index=False)
print("Data cleaned and local backup saved.")

# --- PHASE 3: POSTGRESQL INJECTION ---

db_string = 'postgresql://postgres:fakepassword@localhost:5432/agency_portfolio'

engine = create_engine(db_string)


df_clean.to_sql('creator_finances', engine, if_exists='replace', index=False)

print("Backend Execution Successful! The clean data is officially in PostgreSQL. ðŸš€")
```
